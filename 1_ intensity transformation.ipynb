## 1. Intensity Transformation

    Applies an intensity transformation to an image.
    
    The transformation affects pixel values between 50 and 150, while leaving others unchanged.

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def intensity_transformation(img: np.ndarray) -> np.ndarray:
    new_img = img.copy().astype(np.float32)  # Convert image to float for the transformation

    # Apply the transformation to the relevant range
    mask = (new_img >= 50) & (new_img <= 150)
    new_img[mask] = 1.55 * new_img[mask] + 22.5

    # Clip values to be within valid intensity range [0, 255]
    new_img = np.clip(new_img, 0, 255).astype(np.uint8)

    return new_img


# Load a grayscale image

```
